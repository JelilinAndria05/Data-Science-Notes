## Q.no:1 
The historical average score in a certification exam is 65. A training institute introduces a new course. A random sample of 50 students trained under the new course has a mean score of 68 and a standard deviation of 10. At the 5% level of significance, test whether the new course improves performance.

* **One Sample t-test Right-Tailed**
* **Null Hypothesis (H0)** : mean <= 65
* **Alternate Hypothesis (Ha)** : mean > 65

```python
from scipy import stats
import numpy as np

sample_mean = 68
population_mean = 65
sample_sd = 10
n = 50

t_stat = (sample_mean - population_mean) / (sample_sd / np.sqrt(n))
print("t-value =", t_stat)

critical_value = stats.t.ppf(0.95, df=49)
print("Critical value =", critical_value)

if t_stat > critical_value:
    print("We reject H0")

else:
    print("We do not reject H0")
```

* **Conslusion:**
    At the 5% significance level, there is sufficient statistical evidence to conclude that the new training course improves student performance in the certification exam. This suggests that the institute’s new course is effective and helps students achieve better scores.

## Q.no:3
A company claims that 60% of its users are satisfied with its product. In a survey of 200 users, 134 users report satisfaction. At the 5% level of significance, test whether customer satisfaction is higher than claimed.

* **One sample proportion test right tailed**
* **Null Hypothesis (H0)** : p0 <= 0.60
* **Alternate Hypothesis (Ha)** : p0 > 0.60

```python
from scipy import stats
import numpy as np

p0 = 0.60
n = 200
satisfied_users = 134

p1 = satisfied_users / n
print("Sample proportion =", p1)

z_stat = (p1 - p0) / np.sqrt((p0 * (1 - p0)) / n)
print("Z-value =", z_stat)

critical_value = stats.norm.ppf(0.95)
print("Critical value =", critical_value)

if z_stat > critical_value:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is sufficient statistical evidence to conclude that customer satisfaction is higher than the company’s claimed 60%. This suggests that the product is performing better than expected in terms of user satisfaction, which is a positive sign for the company. It may help improve brand reputation, customer retention, and future business growth.

## Q.no:5 
Two teaching methods are used to train employees. A sample of 30 employees trained using Method A has a mean score of 75 with a standard deviation of 12. A sample of 28 employees trained using Method B has a mean score of 70 with a standard deviation of 10. At the 5% level of significance, test whether there is a difference in average scores between the two methods.

* **Two sample t test two tailed** 
* **Null Hypothesis (H0)** : u1 = u2
* **Alternate Hypothesis (Ha)** : u1 not equal to u2

```python
from scipy import stats
import numpy as np

u1 = 75
sd1 = 12
n1 = 30

u2 = 70
sd2 = 10
n2 = 28

t_stat = (u1 - u2) / np.sqrt((sd1**2 / n1) + (sd2**2 / n2))
print("t-value =", t_stat)

df = n1 + n2 - 2

critical_value = stats.t.ppf(0.975, df=df)
print("Critical value =", critical_value)

if abs(t_stat) > critical_value:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is not enough statistical evidence to conclude that there is a significant difference in average employee scores between Method A and Method B. This means both teaching methods appear to produce similar training outcomes. The company cannot say that one method is significantly better than the other based on this sample.

## Q.no:7 
Two versions of a website are tested. Out of 500 users for Version A, 320 users click the signup button. Out of 480 users for Version B, 290 users click the signup button. At the 5% level of significance, test whether the conversion rates differ

* **Two sample proportion test two tailed** 
* **Null Hypothesis (H0)** : p1 = p2
* **Alternate Hypothesis (Ha)** : p1 not equal to p2

```python
from scipy import stats
import numpy as np

x1 = 320
n1 = 500

x2 = 290
n2 = 480

p1 = x1 / n1
p2 = x2 / n2

print("Sample proportion p1 =", p1)
print("Sample proportion p2 =", p2)

p = (x1 + x2) / (n1 + n2)

z_stat = (p1 - p2) / np.sqrt(p * (1 - p) * ((1/n1) + (1/n2)))
print("Z-value =", z_stat)

critical_value = stats.norm.ppf(0.975)
print("Critical value =", critical_value)

if abs(z_stat) > critical_value:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is not enough statistical evidence to conclude that the conversion rates of Version A and Version B are significantly different. This means both website versions perform similarly in attracting users to click the signup button. Based on this sample, the company cannot say that one version is significantly better than the other for improving conversions.

## Q.no:9 
A company wants to compare the average productivity of employees trained using three different training programs. Program A has 25 employees with a mean productivity score of 80 and a standard deviation of 8. Program B has 30 employees with a mean of 75 and a standard deviation of 7. Program C has 28 employees with a mean of 82 and a standard deviation of 6. At the 5% level of significance, test whether there is a difference in mean productivity across the programs.

* **One way ANOVA**
* **Null Hypothesis (H₀)** : μ1 = μ2 = μ3
* **Alternative Hypothesis (Ha)** : At least one mean is different

```python
from scipy import stats

u1 = 80
sd1 = 8
n1 = 25

u2 = 75
sd2 = 7
n2 = 30

u3 = 82
sd3 = 6
n3 = 28

group1 = stats.norm.rvs(loc=u1, scale=sd1, size=n1)
group2 = stats.norm.rvs(loc=u2, scale=sd2, size=n2)
group3 = stats.norm.rvs(loc=u3, scale=sd3, size=n3)

f_stat, p_value = stats.f_oneway(group1, group2, group3)

print("F-value =", f_stat)
print("P-value =", p_value)

if p_value < 0.05:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is sufficient statistical evidence to conclude that there is a significant difference in mean productivity among the three training programs This means all training programs do not produce the same employee productivity. At least one program performs differently from the others. The company should further analyze which training program gives the best productivity and consider adopting that program more widely.

## Q.no:10 
A company wants to test whether gender and product preference are independent. The survey results are as follows:
●	Male: 60 prefer Product A, 40 prefer Product B
●	Female: 50 prefer Product A, 50 prefer Product B
At the 5% level of significance, test for independence.

* **Chi-Square Test of Independence**
* **Null Hypothesis (H0)** : Gender and product preference are independent
* **Alternative Hypothesis (Ha)**: Gender and product preference are dependent

```python
from scipy.stats import chi2_contingency
import numpy as np

data = np.array([[60, 40],[50, 50]])

chi2, p_value, dof, expected = chi2_contingency(data)

print("Chi-square value =", chi2)
print("P-value =", p_value)
print("Degree of freedom =", dof)

if p_value < 0.05:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is not enough statistical evidence to conclude that gender and product preference are related.
This means gender and product preference appear to be independent. The company cannot say that male and female customers have significantly different preferences between Product A and Product B based on this survey.

## Q.no: 11
A  hospital studies whether smoking status is related to disease occurrence. The data collected is as follows:
●	Smokers: 70 have the disease, 30 do not
●	Non-smokers: 40 have the disease, 60 do not
At the 5% level of significance, test whether smoking status and disease occurrence are independent. 



* **Chi-Square Test of Independence**
* **Null Hypothesis (H0)** : Smoking status and disease occurrence are independent
* **Alternative Hypothesis (Ha)** : Smoking status and disease occurrence are dependent

```python
from scipy.stats import chi2_contingency
import numpy as np

table = np.array([[70, 30],[40, 60]])

chi_value, p_value, df, expected = chi2_contingency(table)

print("Chi-square value =", chi_value)
print("P-value =", p_value)
print("Degree of freedom =", df)

if p_value < 0.05:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is sufficient statistical evidence to conclude that smoking status and disease occurrence are related. This means smoking status and disease occurrence are not independent. The hospital can infer that smokers and non-smokers show significantly different disease occurrence patterns, suggesting that smoking may be associated with a higher risk of disease.

## Q.no: 12
A dice manufacturer claims that a die is fair. The die is rolled 120 times, and the observed frequencies are:
●	Face 1: 18
●	Face 2: 22
●	Face 3: 20
●	Face 4: 25
●	Face 5: 17
●	Face 6: 18
At the 5% level of significance, test whether the die is fair.

* **Chi-Square Goodness of Fit Test**
* **Null Hypothesis (H0)** : The die is fair
* **Alternative Hypothesis (Ha)** : The die is not fair

```python
from scipy.stats import chisquare
import numpy as np

observed = np.array([18, 22, 20, 25, 17, 18])

expected = np.array([20, 20, 20, 20, 20, 20])

chi_value, p_value = chisquare(observed, expected)

print("Chi-square value =", chi_value)
print("P-value =", p_value)

if p_value < 0.05:
    print("We reject H0")
else:
    print("We do not reject H0")
```

* **Conclusion:** At the 5% level of significance, there is not enough statistical evidence to conclude that the die is unfair.
This means the observed frequencies are close enough to what we would expect from a fair die. The dice manufacturer’s claim that the die is fair can be accepted based on this sample data.
