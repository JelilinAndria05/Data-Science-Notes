## A company that the average time taken to resolve customer tickets is 30 minutes. A random sample of 40 tickets shows a mean resolutiom time of 33 minutes. At the 5% level of significance, tet whether the average resolution time has increased.

H0 = Current process / There is no significant difference in the average resolution time
H1 = Alternate process / There is asignificant difference in the average resolution time

```python
from scipy import stats
import numpy as np
```

```python
observations_40_tickets = np.random.normal(loc = 33, scale = 5.0, size = 40).astype("int")
observations_40_tickets
```

```python
t_stats, p_value = stats.ttest_1samp(a = observations_40_tickets, popmean = 30, alternative = "greater")
t_stats, p_value
```

```python
if p_value < 0.05:
    print("we reject H0")
else:
    print("we do not reject H0")
```
