## Dictionary

### Clear

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
items.clear()
items
```

### Copy

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.copy()
print(a)
```

### Fromkeys

```python
items = dict.fromkeys(['orange', 'carrot', 'milk', 'onion', 'garlic'], 1)
items
```

### Get

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.get('orange')
print(a)
```

### Items

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.items()
print(a)
```

### Keys

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.keys()
print(a)
```

### Pop

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
items.pop('orange')
items
```

### Popitem

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.popitem()
print(a)
items
```

### Setdefault

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.setdefault('apple', 5)
print(items)
```

### Update

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.update({'apple':5})
print(items)
```

### Values

```python
items = {'orange':3, 'carrot':2, 'milk':3, 'onion':2.5, 'garlic':1}
a = items.values()
print(a)
```
