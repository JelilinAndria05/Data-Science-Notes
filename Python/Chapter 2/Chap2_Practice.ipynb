# PYTHON DATA STRUCTURES

-> To collect multiple data in a structure

Four Data Structures in Python:-

1. LIST
2. SETS
3. TUPLES
4. DICTIONARIES - Key & Value

## LIST

1. One Dimensional
2. Heterogenous
3. Duplicate Values
4. Indexing & Slicing
5. Mutable

```python
empty_cart1 = []
empty_cart1
```

```python
type(empty_cart1)
```

## SETS

1. Set will allow only unique values. It won't allow duplicate values
2. One Dimensional
3. Mutable
4. Heterogenous
5. Indexing & Slicing not possible


```python
empty_cart2 = set()
empty_cart2
```

```python
type(empty_cart2)
```

## TUPLE

1. One Dimensional
2. Heterogenous
3. Duplicate values
4. Indexing & Slicing
5. Immutable

```python
empty_cart3 = ()
empty_cart3
```

```python
type(empty_cart3)
```

## Dictionary

1. 2 Dimensional
2. Heterogenous
3. Mutable
4. Indexing & Slicing is not possible
5. Don't create duplicate key, but you can create duplicate values

```python
empty_cart4 = {}
empty_cart4
```

```python
type(empty_cart4)
```
