# Why OOPS

## Real world analogy - 1

-> I want to reuse the same function for team members & managers, but they have small differences, How do i do that?

-> Both team members & managers should be able to login. but

* **Team Members** -  Should log in , view tasks & submit tasks.
* **Managers**     - Should lod in, view tasks & assign  tasks

# Without OOPS

```python
def login(user):
    return f"{user['name']} has logged in."

def view_tasks(user):
    return f"{user['name']} can view assigned tasks."
    
def submit_tasks(user):
    if user["role"] == "team_member":
        return f"{user['name']} has submitted the tasks."
    else:
        return "only team members cn submit the tasks."

def assign_tasks(user):
    if user['role'] == 'manager':
        return f"{user['name']} has assigned the tasks."
    else:
        return "Only manager can assign the tasks"
```

```python
# sample users

John  = {'name': 'John', 'role': 'team_member'}
Nasir = {'name': 'Nasir', 'role': 'manager'}
```

```python
# Function calls

print(login(John))
print(view_tasks(John))
print(submit_tasks(John))
print(assign_tasks(John))
print('===================================================================================================')
print('===================================================================================================')
print(login(Nasir))
print(view_tasks(Nasir))
print(assign_tasks(Nasir))
print(submit_tasks(Nasir))
```

## Drawbacks without OOPS:




|       Limitation         |                                                 Explanation                                                             |
|--------------------------|-------------------------------------------------------------------------------------------------------------------------|
|  No reusability of logic |  you can't bundle shared behavior. you repeat logic checks (if role == ... ) again & again                              | 
|  Hard to extend          |  Adding a new role (like admin) means changing all four functions to include new if conitions                           |
|  No structure            |  No way to logically group related functionality. Tasks for team members & managers are spread across multiple functions|
|  No data binding         |  Data (name, role) is not attached to behavior. you have to keep passing the user dict again & again                    |
|  No inheritance          |  You can't inherit shared behavior, Even common methods like login() get repeated logic                                 |

# With OOPS

```python
class Employee:
    def __init__(self,name):
        self.name = name

    def login(self):
        return f"{self.name} has logged in."

    def view_task(self):
        return f"{self.name} can view assigned tasks."

class TeamMember(Employee):
    def submit_task(self):
        return f"{self.name} has submitted the task."
        
class Manager(Employee):
    def assign_task(self):
        return f"{self.name} has assigned a new task."

# Usage

John = TeamMember("John")
Nasir =  Manager("Nasir")

print(John.login())
print(John.submit_task())
print(Nasir.login())
print(Nasir.assign_task())
```

## Benefits of OOPS 

* **With OOPS, the code is clean,stable, structured.**
