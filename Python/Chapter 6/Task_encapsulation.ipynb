# ENCAPSULATION

Encapsulation means we keep some data private and don’t allow direct access.

In this example, the ATM PIN is hidden inside the class.
We can only check it using a method.

```python
class SecureUser:
    def __init__(self, name, atm_pin):
        self.name = name
        self.__atm_pin = atm_pin   # private

    def check_pin(self, attempt):
        if attempt == self.__atm_pin:
            print("PIN is correct")
        else:
            print("PIN is wrong")
```

```python
user = SecureUser("Ram", 1234)

user.check_pin(1234)
user.check_pin(1111)
```
