## Real world analogy - 1

-> I want to reuse the same function for team members & managers, but they have small differences, How do i do that?

-> Both team members & managers should be able to login. but

* **Team Members** -  Should log in , view tasks & submit tasks.
* **Managers**     - Should lod in, view tasks & assign  tasks

```python
# Defining a function

def login(user_data):
    print(f"{user_data['name']} has logged in") 

def view_task(user_data):
    print(f"{user_data['name']} is viewing the task")

def submit_task(user_data):
    if user_data['role'] == "Team_Member":
        print(f"{user_data['name']} has submitted the task")
    else:
        print("Manager no need to submit the task")

def assign_task(user_data):
    if user_data['role'] == "Manager":
        print(f"{user_data['name']} has assigned the task")
    else:
        print("Team Member can't assign the task.")

# Data Collection

user_data = {'name': 'Nasir', 'role': 'Team_Member'}

# Function to be called

login(user_data)
view_task(user_data)
submit_task(user_data)
assign_task(user_data)            ## Now we can understand the limitation  of function - module coding
```

## With OOPS

* **Class** - Blueprint that contains attributes of a particularobjects & its all the functions.
* **Object** - Instance of a class.

```python
class Employee:
    def __init__(self,name):
        self.name = name

    def login(self):
        print(f"{self.name} has logged in")

    def view_task(self):
        print(f"{self.name} has viewing the task")

class TeamMember(Employee):   
    def submit_task(self):
        print(f"{self.name} has submitted the task")

class Manager(Employee):
    def assign_task(self):
        print(f"{self.name} has assigned the task")
```

```python
John = TeamMember("John")
Nasir = Manager("Nasir")
```

```python
John.login()
John.submit_task()
John.view_task()
Nasir.login()
Nasir.assign_task()
```
