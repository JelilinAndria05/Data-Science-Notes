# ABSTRACTION

Abstraction means we hide how things work inside and only show what is needed to the user.

In this example, the user only enters the PIN and tries to access the account.
They don't know what is happening inside the program.

```python
class SecureUser:
    def __init__(self, name, atm_pin):
        self.name = name
        self.__atm_pin = atm_pin

    def access_account(self, attempt):
        print("Checking PIN...")
        if attempt == self.__atm_pin:
            print("Access Granted")
        else:
            print("Wrong PIN")
```

```python
user = SecureUser("Ram", 1234)
user.access_account(1234)
user.access_account(1111)
```

# INHERITANCE

Inheritance means we create a new class using an already existing class.

In this example, PremiumUser is created from SecureUser, so it can use its features.

```python
class PremiumUser(SecureUser):
    def loan_facility(self):
        print("Loan facility available for premium user")
```

```python
user = PremiumUser("John", 5678)
user.loan_facility()
```

# POLYMORPHISM

Polymorphism means one method can work in different ways.

Here, the same withdrawal method behaves differently for normal user and premium user.

```python
class SecureUser:
    def cash_withdrawal(self):
        print("Normal withdrawal")

class PremiumUser(SecureUser):
    def cash_withdrawal(self):
        print("High limit withdrawal for premium user")
```

```python
user1 = SecureUser()
user2 = PremiumUser()

user1.cash_withdrawal()
user2.cash_withdrawal()
```
