## 1. Groupby

```python
import pandas as pd
```

```python
insurance_data = pd.read_csv("insurance.csv")
insurance_data
```

### 1.1 Perform initial analysis

```python
insurance_data.isna().sum()
```

```python
insurance_data.dtypes
```

```python
insurance_data.describe()
```

### 1.2 Pull up the insurance charges based on region 

```python
insurance_data["region"].unique()
```

```python
insurance_data["region"].nunique()
```

```python
insurance_data.groupby(by = "region")["charges"].mean()
```

```python
insurance_data.groupby(by = "region")["charges"].mean().round(2)
```

```python
insurance_data.groupby(by = "region")["charges"].mean().round(2).sort_values()
```

### 1.3 Pull up the insurance charges based on region & sex

```python
insurance_data.groupby(by = ["region", "sex"])["charges"].mean().round(2).sort_values()
```

```python
insurance_data.groupby(by = ["region", "sex"])["charges"].mean().round(2).sort_values(ascending = False)
```

## 2. Pivot_table

* Prsentable comparing 2 categorical features Vs 1 continuous

=> Continuous - countable/segmentable/has unit

=> Discrete   - countable/ no unit

```python
pd.pivot_table(data = insurance_data, values = "charges", index = "region", columns = "sex").round()
```

```python
import pandas as pd
pd.pivot_table(data = insurance_data, values = "charges", index = "region", columns = ["sex", "children"]).round()
```

## 3. Aggregate

```python
insurance_data.head()
```

### 3.1 Find out the average insurance charges based on region 

```python
insurance_data.groupby(by = "region")["charges"].mean().round(2).sort_values(ascending = False)
```

```python
insurance_data.groupby(by = ["region", "sex"])["charges"].mean().round(2).sort_values(ascending = False)
```

```python
insurance_data.groupby(by = "region")["charges"].aggregate(func = ["mean", "min", "max", "sum"]).round()
```

```python
insurance_data.groupby(by = ["region", "sex"])["charges"].aggregate(func = ["mean", "min", "max", "sum"]).round()
```

## 5. Concat

```python
sales_data_2017 = pd.read_csv("Sales Transcations-2017.csv")
sales_data_2018 = pd.read_csv("Sales Transcations-2018.csv")
sales_data_2019 = pd.read_csv("Sales Transcations-2019.csv")
```

## 6. Apply

```python
insurance_data
```

### 6.1 Create a new column in the same data frame with the name "Eligibility Criteria" where if age is > 20, say eligible else not

```python
def get_age(x):
    if x > 20:
        return "Eligible"
    else:
        return "Not Eligible"
```

```python
insurance_data["Eligibility Criteria"] = insurance_data["age"].apply(get_age)
insurance_data
```

## 7. Lambda

* Also called as Anonymous Function / Nameless Function


```python
insurance_data["Eligibility Criteria_2"] = insurance_data["age"].apply(lambda x:1 if x > 20 else 0)
insurance_data
```
