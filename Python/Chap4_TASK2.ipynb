# TASK 1

```python
import pandas as pd
```

```python
sales_data_2017 = pd.read_csv("Sales Transcations-2017.csv")
sales_data_2018 = pd.read_csv("Sales Transcations-2018.csv")
sales_data_2019 = pd.read_csv("Sales Transcations-2019.csv")
```

```python
sales_full_data = pd.concat([sales_data_2017, sales_data_2018, sales_data_2019])
sales_full_data
```

### Performing initial analysis

```python
sales_full_data.dtypes
```

```python
sales_full_data.isna().sum()
```

```python
sales_full_data.shape
```

```python
sales_full_data.describe()
```

### Lets check for duplicated data

```python
sales_full_data.duplicated().sum()
```

```python
sales_full_data[sales_full_data.duplicated()].head(10)
```

```python
sales_full_data = sales_full_data.drop_duplicates()
```

```python
sales_full_data.duplicated().sum()
```

### Lets converts Qty & Rate from object to numeric

```python
sales_full_data.loc[:, 'Qty'] = sales_full_data['Qty'].astype(int)
sales_full_data
```

```python
sales_full_data.loc[:, "Qty"] = pd.to_numeric(sales_full_data["Qty"], errors = 'coerce')
sales_full_data
```

```python
sales_full_data.loc[:, 'Rate'] = pd.to_numeric(sales_full_data["Rate"], errors = 'coerce')
sales_full_data
```

```python
sales_full_data.dtypes
```

### Lets calculate the gross

```python
sales_full_data.loc[:, "Gross"] = sales_full_data["Qty"] * sales_full_data["Rate"]
sales_full_data
```

```python
sales_full_data.dtypes
```

### Performing to_datatime

```python
sales_full_data.loc[:, 'Date'] = pd.to_datetime(sales_full_data['Date'], dayfirst=True, errors='coerce')
sales_full_data
```

```python
sales_full_data.dtypes
```

### Lets remove the missing datas

```python
sales_full_data = sales_full_data.dropna(subset = ['Qty'])
sales_full_data
```

```python
sales_full_data = sales_full_data.dropna(subset=['Date'])
sales_full_data
```

```python
sales_full_data = sales_full_data.dropna(subset=['Rate'])
sales_full_data
```

### Lets clean the discount

```python
sales_full_data.loc[:, 'Disc'] = sales_full_data['Disc'].fillna(0)
sales_full_data
```

```python
sales_full_data.loc[:, 'Disc'] = pd.to_numeric(sales_full_data['Disc'], errors='coerce')
sales_full_data
```

### Lets check voucher

```python
sales_full_data.loc[:, 'Voucher Amount'] = (sales_full_data.groupby('Voucher')['Gross'].transform('sum'))
sales_full_data
```

### Final Check

```python
sales_full_data.isna().sum()
```

```python
sales_full_data.dtypes
```

```python
sales_full_data.describe()
```

# TASK 2

* Merge Function

```python
import pandas as pd

data1 = { "Product_ID": [1, 2, 3], "Product_Name": ["Pen", "Pencil", "Book"] }

data1 = pd.DataFrame(data1)
data1
```

```python
data2 = { "Product_ID": [1, 2, 4], "Price": [10, 5, 100] }

data2 = pd.DataFrame(data2)
data2
```

## Inner

```python
pd.merge(data1, data2, on='Product_ID', how='inner')
```

## Left

```python
pd.merge(data1, data2, on='Product_ID', how='left')
```

## Right

```python
pd.merge(data1, data2, on='Product_ID', how='right')
```

## Outer

```python
pd.merge(data1, data2, on='Product_ID', how='outer')
```

* List Comprehension

```python
numbers = [x**2 for x in range(6)]

print(numbers)
```

```python
even_numbers = [x for x in range(1,11) if x % 2 == 0]

print(even_numbers)
```

# Task 4

* Lambda Function

## Map

```python
numbers = [1, 2, 3, 4]

result = list(map(lambda x: x**2, numbers))

print(result)
```

## Filter

```python
numbers = [1,2,3,4,5,6]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```

## Reduce

```python
from functools import reduce

numbers = [1,2,3,4]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```
