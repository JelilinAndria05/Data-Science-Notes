#### Parameters & Arguments:-

```python
def addition(first_name, second_name, third_name):
    return first_name+  second_name+ third_name
```

```python
addition(10,20,30)
```

#### Positional Arguments:-

```python
def get_details(name, place, price):
    print("Client Name: ", name)
    print("Client Place: ", place)
    print("Client Price: ", price)
```

```python
get_details(place = "Chennai", price = "50,000", name = "Jello")
```

#### Default Arguments & Non Default Arguments:-

```python
def get_details_Chennai(name, price, place = "Chennai"):
    print("Client Name: ", name)
    print("Client Place: ", place)
    print("Client Price: ", price)
```

```python
get_details_Chennai(price = 50, name = "Jelilin")
```

#### Variable Argument:-

```python
def get_details(name, place, *price):
    print("Client Name: ", name)
    print("Client Place: ", place)
    print("Client Price: ", price)
```

```python
get_details("Jello", "Chennai", 0,60,70)
```

#### Keyboard Argument:-

```python
def get_details(name, place, **price):
    print("Client Name: ", name)
    print("Client Place: ", place)
    print("Client Price: ", price)
```

```python
get_details("Jello", "Chennai", ask_1 = 90, ask_2 = 80, ask_3 = 70)
```

### 5.4 Challenge to find the factors of a number

```python
user_input = int(input("Enter your number: "))
result = []
for i in range(1, user_input+1):
    if user_input %i == 0:
        result.append(i)
print("The factors of {} is {}.".format(user_input,result))
```
