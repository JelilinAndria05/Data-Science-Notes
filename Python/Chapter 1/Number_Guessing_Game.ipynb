```python
def Number_Guessing_Game():
    import random
    print("Welcome to the Number Guessing Game")
    print("Guess the Number between 1 to 100")
    secret_number=random.randint(1,100)
    for i in range(1,15):
        try:
            user_input=int(input("Enter the Number: "))
            if user_input < 1 or user_input > 100:
                print("Invalid Number")
            elif user_input < secret_number:
                print("Your Number is Lesser")
            elif user_input > secret_number:
                print("Your Number is Higher")
            else:
                print("Your Guess is CORRECT")
                break
        except ValueError:
            print("Invalid Characters, Enter between 1 to 100 to play this Game")
    if secret_number==user_input:
        print(f"Congrats! You have guessed the number {secret_number} in {i} attempts")
    else:
        print("Sorry! All your trails is exceeded")
```

```python
Number_Guessing_Game()
```
