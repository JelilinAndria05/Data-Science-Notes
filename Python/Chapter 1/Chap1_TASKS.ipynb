# TASK 3
=>Rock, Paper, Scissors Game

```python
def Rock_Paper_Scissors():
    import random
    print("Welcome to the Rock, Paper, Scissors Game")
    print("Choose Rock or Paper or Scissors.\n")

    choices = ("Rock", "Paper", "Scissors")

    while True:
        try:
            computer = random.choice(choices)
            user = input("Enter Rock, Paper or Scissors: ").capitalize()

            if user in choices:
               
                print("Computer chose:", computer)

                if user == computer:
                    print("It's a DRAW.\n")

                elif (user == "Scissors" and computer == "Paper") or \
                      (user == "Rock" and computer == "Scissors") or \
                      (user == "Paper" and computer == "Rock"):
                    print("YOU WIN.\n")

                else:
                    print("Computer WINS.\n")

                print("Do you want to play again?")
                rematch = input("Enter Yes or No: ").capitalize() 

                if rematch == "Yes":
                    print()
                    continue
                else:
                    break
            else:
                print("INVALID INPUT.\n")
        except:
            print("Something went wrong!.\n")
```

# TASK 4
=>Word Guessing Game (Mini Hangman)

```python
import random

words_list = ["python", "machine", "bridge", "learning"]
print("Welcome to Mini Hangman Game!!!")
print("You have 6 chances to guess the word.\n")

word = random.choice(words_list)
guessed_word = ["_"] * len(word)
attempts = 6
guessed_letters = []

while attempts > 0 and "_" in guessed_word:

    print("word:", " ".join(guessed_word))
    print("Wrong attempts left: ",attempts)

    guess = input("Guess a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("Please Enter only one alphabet letter.\n")
        attempts -= 1
        continue

    if guess in guessed_letters:
        print("You already guessed this letter! Attempt reduced.\n")
        attempts -= 1
        continue

    guessed_letters.append(guess)

    
    if guess in word:
        print("CORRECT Guess!\n")

        for i in range(len(word)):
            if word[i] == guess:
                guessed_word[i] = guess

    else:
        print("WRONG Guess\n")
        attempts -=1

if "_" not in guessed_word:
    print("Congratulations! You guessed the word: ")
else:
    print("Game Over! The word was: ",word)
```
