## 1. Why is Python preferred in the field of Data Science ?

    Python is widely used in Data Science because it is easy to learn and has simple syntax. It provides powerful libraries such as NumPy, Pandas, and Matplotlib for data analysis and visualization. Python also has strong community support and can handle large amounts of data efficiently.

## 2. What is the difference between a Jupyter Notebook and Google Colab?

    Jupyter Notebook is an open-source tool that runs on a local computer to write and execute Python code. Google Colab is a cloud-based platform that runs in the browser and does not require installation. Colab also provides free GPU and easy sharing through Google Drive.

## 3. When would you use a list over a set or a tuple?

    A list is used when we need an ordered collection of items that can be modified. Lists allow duplicate values and support operations like adding, removing, and updating elements. Sets remove duplicates, while tuples are immutable (cannot be changed).

## 4. What is exception handling? Give a real-life example.

    Exception handling is used to handle errors during program execution without stopping the program. It uses try and except blocks to manage errors.
    Example: When dividing a number by zero, the program can show an error message instead of crashing.

## 5. Define DataFrame and Series in Pandas. What’s the difference?

    A Series is a one-dimensional labeled array that can store a single column of data. A DataFrame is a two-dimensional table with rows and columns. The main difference is that a Series contains one column, while a DataFrame contains multiple columns.

## 6. What are the different ways of dealing with missing values in a dataset?

    Missing values can be handled by removing rows or columns with missing data. They can also be replaced with mean, median, or mode values. Another method is filling them with a constant value or using interpolation techniques.

## 7. How do NumPy and Pandas differ in terms of functionality?

    NumPy is mainly used for numerical computations and working with arrays. Pandas is used for data analysis and handling structured data such as tables. Pandas also provides DataFrame and Series structures, which are not available in NumPy.

## 8. What is the role of libraries like Seaborn and Matplotlib in a data project?

    Seaborn and Matplotlib are used for data visualization. They help in creating graphs, charts, and plots to understand patterns in data. These visualizations make it easier to analyze and present data results.

## 9. When should you use .loc[] and when .iloc[]?

    .loc[] is used to access data using row and column labels. 
    .iloc[] is used to access data using numerical index positions. 
    In simple terms, .loc[] works with names, while .iloc[] works with numbers.

## 10. What is the importance of writing functions in Python?

    Functions in Python help organize code into small and reusable blocks. They reduce code repetition and make programs easier to read and maintain. Functions also help in debugging and improving program efficiency.

## 9. Replace all missing values in a DataFrame with the column's mean

```python
import pandas as pd
import numpy as np
data = { "marks": [80, 90, np.nan, 70],
         "age": [20, np.nan, 22, 21] }
data = pd.DataFrame(data)
data_updated = data.fillna(data.mean())
print(data_updated)
```
