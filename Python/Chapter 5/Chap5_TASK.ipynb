## Survival Count

```python
import seaborn as sns

sns.countplot(data = titanic_data, x = "survived", y = None)
plt.show()
```

```python
titanic_data["survived"].value_counts()
```

## Survival based on Gender

```python
sns.barplot(data = titanic_data, x = "sex", y = "survived")
plt.show()
```

## Survival based on Class

```python
sns.barplot(data = titanic_data, x = "pclass", y = "survived")
plt.show()
```

## Age Distribution

```python
sns.histplot(data=titanic_data, x="age")
plt.title("Age Distribution")
plt.show()
```

## Embarkation Location

```python
sns.countplot(data=titanic_data, x="embarked")
plt.title("Passengers by Embark Location")
plt.show()
```

## Family aboard (Alone vs Not Alone)

## Who column vs Survival
